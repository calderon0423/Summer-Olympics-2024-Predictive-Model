## Preprocess the data

```python
# Split the data into training and testing sets using random_state=1

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Encode y data with Label Encoder. 

le = LabelEncoder()

y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.fit_transform(y_test)
```

```python
# Encode X data with Ordinal Encoder. 

#? X_train_encodered_discipl_event_gender = ohe.fit_transform(X_train['Discipline', 'Event', 'Gender']
#? X_test_encodered_discipl_event_gender = ohe.fit_transform(X_test['Discipline', 'Event', 'Gender']                                                      
#? oe = OrdinalEncoder()
#? X_train_encodered_country = oe.fit_transform(X_train['country']    
#? X_test_encodered_country = oe.fit_transform(X_test['country']

# encode_education = OrdinalEncoder(categories=[['primary', 'secondary', 'tertiary']], handle_unknown='use_encoded_value', unknown_value=-1)
# encode_education.fit(X_train_filled['education'].values.reshape(-1, 1))

oe = OrdinalEncoder()

X_train_encoded = oe.fit_transform(X_train)
X_test_encoded = oe.fit_transform(X_test)

X_train_encoded
```

```python
# Scale the X data by using StandardScaler()

scaler = StandardScaler().fit(X_train)

X_train_scaled = scaler.transform(X_train_encoded)
X_train_scaled
```

```python
# Transform the test dataset based on the fit from the training dataset

X_test_scaled = scaler.transform(X_test_encoded)
X_test_scaled
```

```python
# Scale y data

y_train_scaled = scaler.transform(y_train_encoded)
y_test_scaled = scaler.transform(y_test_encoded)

y_train_scaled
```

## Model and Fit to a Logistic Regression Classifier

```python
# Create the logistic regression classifier model with a random_state of 1

lr_model = LogisticRegression()

# Fit the model to the training data

lr_model.fit(X_train_scaled, y_train_scaled)
```

## Model and Fit to a KNN Model

```python
# Loop k to find the best number of neighbors

import matplotlib.pyplot as plt
# Loop through different k values to find which has the highest accuracy.
# Note: We use only odd numbers because we don't want any ties.
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train_scaled)
    train_score = knn.score(X_train_scaled, y_train_scaled)
    test_score = knn.score(X_test_scaled, y_test_scaled)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
# Plot the results
plt.plot(range(1, 20, 2), train_scores, marker='o', label="training scores")
plt.plot(range(1, 20, 2), test_scores, marker="x", label="testing scores")
plt.xlabel("k neighbors")
plt.ylabel("accuracy score")
plt.legend()
plt.show()
```
